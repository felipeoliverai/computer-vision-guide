### Pillow Framework 
<br>

Pillow it means Python Image Library, the most popular and facto standard library python for loading and working with image data is Pillow. 

Pillow is an updated version of the Python Image Library or PIL, and supports a range of simples and sophisticated image manipulation functionality. It is also the basis for simples image support in other Python libraries such as Scipy and Matplotlib. 


<br>
<hr>

```
import os 
import matplotlib.pyplot as plt
from matplotlib import image
import PIL
from PIL import Image


%matplotlib inline 
import warnings 
warnings.filterwarnings("ignore")


print("Pillow version: ", PIL.__version__)
```

#### Load images 



```
# images path
img_1 = "/content/drive/My Drive/Deep Learning - Projetos/computer-vision-guide/images /mountain.jpg"
img_2 = "/content/drive/My Drive/Deep Learning - Projetos/computer-vision-guide/images /bird-blue.jpg"
```

#### Change Image channels 
<br>

```
# change channels of image (RBG ---> GrayScale)
image = Image.open(img_2)

gray_scale = image.convert(mode="L")
gray_scale.save("bird-gray.jpg")
```

```
# visualize different images 

img_rgb = Image.open(img_2)
img_gray = Image.open("bird-gray.jpg")

plt.figure(figsize=(12,6))

# RGB 
plt.subplot(1,2,1)
plt.title("RGB bird", fontsize=15)
plt.imshow(img_rgb)
plt.xticks([])
plt.yticks([])

# Gray scale 
plt.subplot(1,2,2)
plt.title("Gray scale bird", fontsize=15)
plt.imshow(img_gray)
plt.xticks([])
plt.yticks([])

plt.tight_layout()
```

<hr>
<br>


#### Resize Images 

<br>

```
# Resize images 

img = Image.open(img_2)
print("Original size:", img.size)

img.thumbnail((300, 300))
print("Resize:", img.size)
```

```
# Resize image ignore proportion

img = Image.open(img_2)
print("Original size:", img.size)


img_resized = img.resize((350, 350))
print("Resize ignore proportion:", img_resized.size)
```

```
# resize ignore proportion 
plt.imshow(img_resized)
```

<hr>
<br>

#### Manipulation - Flip, Rotate and Crop images

<br>

```
# manipulation 
img = Image.open(img_2)

# horizontal flip 
hor_flip = img.transpose(Image.FLIP_LEFT_RIGHT)


# vertical flip
ver_flip = img.transpose(Image.FLIP_TOP_BOTTOM)



plt.figure(figsize=(12,6))

# Horizontal flip
plt.subplot(1,2,1)
plt.title("Horizontal", fontsize=15)
plt.imshow(hor_flip)
plt.xticks([])
plt.yticks([])


# Vertical Flip 
plt.subplot(1,2,2)
plt.title("Vertical", fontsize=15)
plt.imshow(ver_flip)
plt.xticks([])
plt.yticks([])

plt.tight_layout()
```

<hr>
<br>

#### Rotate image 
<br>

```
# Rotate 

img = Image.open(img_2)
img_rotate = img.rotate(angle=30)


# show rotation
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.title("Original", fontsize=15)
plt.xticks([])
plt.yticks([])
plt.imshow(img)




plt.subplot(1,2,2)
plt.title("Rotate", fontsize=15)
plt.xticks([])
plt.yticks([])
plt.imshow(img_rotate)


plt.tight_layout()
```

<hr>
<br>

#### Cropped 
<br>

```
# Cropped Image (top, lef, bottom, right)

img = Image.open(img_2)
cropped = img.crop(box=(150, 150, 400, 400))

print("Cropped image:", cropped.size)

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.title("Original", fontsize=15)
plt.xticks([])
plt.yticks([])
plt.imshow(img)


plt.subplot(1,2,2)
plt.title("Cropped", fontsize=15)
plt.xticks([])
plt.yticks([])
plt.imshow(cropped)

plt.tight_layout()
```
